```python
import pandas as pd

df = pd.read_csv("../data/delhi_weather_2025.csv.csv")

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['date_ist'] = pd.to_datetime(df['date_ist'])
```

```python
df['date_ist'].dtype
```

```python
df['month'] = df['date_ist'].dt.month
```

```python
df[['date_ist','month']].head()
```

```python
monthly_temp = df.groupby('month')['temp_c'].mean()
monthly_temp
```

```python
import matplotlib.pyplot as plt

plt.figure()
monthly_temp.plot()

plt.title("Average Monthly Temperature - Delhi 2025")
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")

plt.show()
```

```python
# Hottest day
hottest_day = df.loc[df['temp_c'].idxmax()]
hottest_day
```

```python
# Coldest day
coldest_day = df.loc[df['temp_c'].idxmin()]
coldest_day
```

```python
yearly_avg_temp = df['temp_c'].mean()
yearly_avg_temp
```

```python
plt.figure()
plt.scatter(df['temp_c'], df['aqi_index'])
plt.xlabel("Temperature (°C)")
plt.ylabel("AQI Index")
plt.title("Temperature vs AQI Relationship")
plt.show()
```

```python
monthly_aqi = df.groupby('month')['aqi_index'].mean()

plt.figure()
monthly_aqi.plot()
plt.title("Average Monthly AQI - Delhi 2025")
plt.xlabel("Month")
plt.ylabel("AQI Index")
plt.show()
```

```python
monthly_humidity = df.groupby('month')['humidity'].mean()

plt.figure()
monthly_humidity.plot()
plt.title("Average Monthly Humidity - Delhi 2025")
plt.xlabel("Month")
plt.ylabel("Humidity (%)")
plt.show()
```
